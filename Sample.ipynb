```python
import PyICER as pic
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pill = pic.cvs_lead('u_pill')
```

```python
def sample_function(x):
    a = 0.349
    return (0.1*x)**a

for l in ['10d','DIARRHEA','APPETITE','Acid base']:
    print l+' :'
    pic.dist_analyze(pic.column_operator(pill,l,sample_function))
```

```python
lst = ['10d','1m','1y']
for i in range(3):
    for j in range(i+1,3):
        pic.pval_paired(pill,lst[i],lst[j])
        
```
